## Graficos:

Edad de usuarios:

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_usuarios['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Edad de Usuarios')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(x=df_usuarios['age'], color='lightgreen')
plt.title('Boxplot de Edad de Usuarios')
plt.show()
```

Usuarios por ciudad:

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df_usuarios['city'], order=df_usuarios['city'].value_counts().index, palette='viridis')
plt.title('Distribución de Usuarios por Ciudad')
plt.xlabel('Cantidad de Usuarios')
plt.ylabel('Ciudad')
plt.show()
```

Duracion de llamadas:

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_calls['duration'], bins=30, kde=True, color='salmon')
plt.title('Distribución de Duración de Llamadas')
plt.xlabel('Duración (minutos)')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(x=df_calls['duration'], color='orchid')
plt.title('Boxplot de Duración de Llamadas')
plt.xlabel('Duración (minutos)')
plt.show()
```

Mensajes por usuario:

```python
msgs_per_user = df_messages.groupby('user_id').size()
plt.figure(figsize=(10, 6))
sns.histplot(msgs_per_user, bins=20, color='teal')
plt.title('Distribución de Número de Mensajes por Usuario')
plt.xlabel('Cantidad de Mensajes')
plt.ylabel('Cantidad de Usuarios')
plt.show()
```

MG usados:

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_internet['mb_used'], bins=30, kde=True, color='gold')
plt.title('Distribución de MB Usados en Internet')
plt.xlabel('MB Usados')
plt.ylabel('Frecuencia')
plt.show()
```
